```python
# Python Code to Generate a WordCloud 
  
# Importing the Libraries 
from wordcloud import WordCloud, STOPWORDS
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# Read 'Youtube05-Shakira.csv' file  
data = pd.read_csv(r"1.csv", encoding ="latin-1")
```

```python
#Setting the comment and stop words
comment_words = '' 
stop_words = set(STOPWORDS)
```

```python
# Iterating through the .csv data file 
for i in data: 
    i = str(i) 
    seperate = i.split() 
    for j in range(len(seperate)): 
        seperate[j] = seperate[j].lower() 
      
    comment_words += " ".join(seperate)+" "
```

```python
# Create the Word Cloud
final_wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='black', 
                stopwords = stop_words, 
                min_font_size = 10).generate(comment_words)
```

```python
# Plotting the WordCloud                    
plt.figure(figsize = (10, 10), facecolor = None) 
plt.imshow(final_wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show()
```
